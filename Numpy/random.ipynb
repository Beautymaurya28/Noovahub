```python
#numpy.random lets you generate random numbers, arrays, and even choices from data.
import numpy as np

x = np.random.randint(100)   # integer from 0 to 99
print(x)
x1 = np.random.rand() #btw 0 to 1
print(x1)
```

```python
#Generate random array
arr = np.random.randint(100, size=5)
print(arr)
arr1 = np.random.randint(100, size=(3, 4))
print(arr1)
```

```python
#random floats
arr = np.random.rand(3, 2)
print(arr) #raange is always 0 to 1
```

```python
#Random choice from a list
arr = np.array([10, 20, 30, 40, 50])
choice = np.random.choice(arr)
print(choice)
#also pick multiple random element
choices = np.random.choice(arr, size=3)
print(choices)

```

```python
#Shuffle (reorder array randomly)
arr = np.array([1, 2, 3, 4, 5])
np.random.shuffle(arr)
print(arr)
#shuffle() changes the original array
```

```python
#Permutation (random reorder copy)
arr = np.array([1, 2, 3, 4, 5])
new_arr = np.random.permutation(arr)
print(new_arr)
#Doesnâ€™t affect the original array.
```

```python
# Random from Normal Distribution
arr = np.random.normal(0, 1, 5)  # mean=0, std=1, size=5
print(arr)
```

```python
# Random Seed (to repeat results)
#same random output every time
np.random.seed(42)
print(np.random.randint(100, size=5))
```
