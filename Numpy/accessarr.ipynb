```python
#how to access the array element
#indexes in numpy array always startwith zero

import numpy as np
beauty=np.array([1,2,3,4])
print(beauty[0])
print(beauty[2])
print(beauty[2]+beauty[3])
```

```python
#access two d-int his we will use comma sepearting indexes
beauty=np.array([[1,2,3,4],[4,5,6,7]])
print(beauty[0,1])
print(beauty[0,3])

```

```python
#accessing 3d
beauty=np.array([[[1,2,3],[1,4,5]],[[7,8,9],[4,7,0]]])
print(beauty[0])
```

```python
#how to access element form negative indexing
beauty=np.array([[1,2,3,4,5],[1,2,7,8,9]])
print(beauty[1,-1])
```
