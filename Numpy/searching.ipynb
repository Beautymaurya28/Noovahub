```python
#searching in array
#where
import numpy as np
a=np.array([1,2,3,4,5,4,4])
b=np.where(a==4)
print(b)
```

```python
#where index is even
a=np.array([1,2,3,4,5,4,4,0])
b=np.where(a%2==0)
print(b)
```

```python
#where index is odd
a=np.array([1,2,3,4,5,4,4,0])
b=np.where(a%2==1)
print(b)
```

```python
#search sorted method searchsorted() by default it is left to right
#where index is even
a=np.array([1,2,3,4,5,4,4,0])
b=np.searchsorted(a,4)
print(b)
```

```python
#search from right to left
a=np.array([1,2,3,4,5,4,4,0])
b=np.searchsorted(a,4,side='right')
print(b)
```

```python
#search from right to left
a=np.array([1,2,3,4,5,4,4,0])
b=np.searchsorted(a,[2,4,6])
print(b)
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
x = np.where(arr == 30)
print(x)
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])
x = np.where(arr % 2 == 0)
print(x)
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])
even_values = arr[np.where(arr % 2 == 0)]
print(even_values)
```

```python
arr = np.array([10, 20, 30, 40, 50])
x = np.searchsorted(arr, 35)
print(x)
```

```python
arr = np.array([10, 20, 30, 40, 50])
x = np.searchsorted(arr, [25, 45])
print(x)
```

```python
arr = np.array([10, 20, 5, 40, 15])
print("Max index:", np.argmax(arr))
print("Min index:", np.argmin(arr))
```
