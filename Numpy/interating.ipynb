```python
#in two d
import numpy as np
beauty=np.array([[1,2,3,4,5],[1,2,3,7,8]])
for x in beauty:
    print(x)
```

```python
#iterating  in 1d
import numpy as np
beauty=np.array([1,2,3,4,5])
for x in beauty:
    print(x)
```

```python
#in 3d
beauty=np.array([[[1,2,3],[3,6,7]],[[5,6,7],[8,7,9]]])
for x in beauty:
    print(x)
```

```python
#iterate in each sacle element of the 2d
beauty=np.array([[[1,2,3],[3,6,7]],[[5,6,7],[8,7,9]]])
for x in beauty:
    for y in x:
        for z in y:
            print(z)
```

```python
#another way: by using nditer()
#itis a helping function for advanced iteration
#iterate through 3d array
beauty=np.array([[[1,2,3],[3,6,7]],[[5,6,7],[8,7,9]]])
for x in np.nditer(beauty):
    print(x)



#.ndenumerator()-getindex+value
for x in np.ndenumerate(beauty):
    print(x)


#modify while iterating
#op_flag['readwrite']
for x in np.nditer(beauty,op_flags=['readwrite']):
    x[...]=x*2
print(beauty)




```
