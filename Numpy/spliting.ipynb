```python
#access the split array as per you want
import numpy as np
a = np.array([1,2,3,4])
b=np.array_split(a,3) #divide the array into 3 parts
print(b[0])
```

```python
#split the 2d into3d as per row
a=np.array([[1,2,6],[3,4,8],[4,5,6],[5,7,8],[1,2,3],[8,9,0]])
b=np.array_split(a,3,axis=1)
print(b)
```

```python
#use hslpit or hstack method to split 2d array into three 2d array along
a=np.array([[1,2,6],[3,4,8],[4,5,6],[5,7,8],[1,2,3],[8,9,0]])
b=np.hsplit(a,3)
print(b)
```
