```python
#numpy data type
#like i for int,f for float,boolean b
#the numpy array object has a property called dtype that return the data type of an object
import numpy as np
# beauty=np.array([1,2,3])
# print(beauty.dtype)
beauty=np.array([1,2,3],dtype='S')
print(beauty)
beauty=np.array([1,2,3],dtype='i4')
print(beauty)
beauty=np.array(['beauty','maurya'])
print(beauty.dtype)
```

```python
#value error
beauty=np.array(['a','2'],dtype='i')
print(beauty) #value error we can convert numerical to string but string can't convert into numerical
```

```python
#converting data type on an existing array-astype()
#example- change data type from float to integer by  using 'i'
beauty=np.array([1.4,2.5,6.8])
beautycopy=beauty.astype('i')
print(beautycopy)
```

```python
#change data type from integer to boolean
beauty=np.array([1,2,3,0])
beautycopy=beauty.astype(bool)
print(beautycopy)
```
