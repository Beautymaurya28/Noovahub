```python
#filter array-boolean index based
#extracting elements from an array that meet a specific condition

import numpy as np

arr = np.array([10, 20, 30, 40, 50])

filter_arr = arr > 25  # condition
print(filter_arr)
new_arr = arr[filter_arr]
print(new_arr)

```

```python
arr = np.array([10, 20, 30, 40, 50])
new_arr = arr[arr > 25]
print(new_arr)
```

```python
arr = np.array([10, 20, 30, 40, 50, 60])

# Get numbers > 20 and < 50
new_arr = arr[(arr > 20) & (arr < 50)]
print(new_arr)
```

```python
arr = np.array(["apple", "banana", "cherry", "mango"])

# Keep only fruits that contain 'a'
new_arr = arr[np.char.find(arr, 'a') >= 0]
print(new_arr)
```

```python
marks = np.array([55, 70, 35, 80, 90])
passed = marks[marks >= 60]

print("Passed Students Marks:", passed)
```
