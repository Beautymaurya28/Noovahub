```python
#puting elementin order sequence
#sort()
#but it doesn't the changeanything in original array
import numpy as np
arr = np.array([0, 1, 0, 2, 3, 0])
print(np.sort(arr))
print(arr)
```

```python
beauty=np.array([3,2,4,5])
print(np.sort(beauty)) #ascending order by default
```

```python
#sort in descending order
beauty=np.array([3,2,4,5])
print(np.sort(beauty)[::-1])
```

```python
#sort 2 d array
arr = np.array([[3, 2, 1], [6, 5, 4]])
print(np.sort(arr))
```

```python
arr = np.array([[3, 2, 1], [6, 5, 4]])
print(np.sort(arr)) #default axis is -1(row)
```

```python
#sort by col
arr = np.array([[3, 2, 1], [6, 5, 4]])
print(np.sort(arr, axis=0))
```

```python
#get the index of sorted array as output
arr = np.array([[3, 2, 1], [6, 5, 4]])
print(np.argsort(arr, axis=0))
arr = np.array([40, 10, 20, 30])
indexes = np.argsort(arr)
print(indexes)

```

```python
names = np.array(["D", "A", "C", "B"])
scores = np.array([40, 10, 20, 30])

sorted_names = names[np.argsort(scores)]
print(sorted_names)
```
